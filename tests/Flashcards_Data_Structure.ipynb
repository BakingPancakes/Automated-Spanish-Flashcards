```python
df3 = pd.DataFrame(data3)
df3.head()
```

```python
df3_exploded = df3.explode(['subword','type','details'])
df3_exploded.head()
```

```python
pd.json_normalize(df3_exploded['details']).head()
```

```python
pd.json_normalize(df3_exploded['details'],'subdefinitions').head()
```

```python
# Reset the index of df3_exploded
df3_exploded.reset_index(drop=True, inplace=True) #! note how indexes of exploded are all 0

# Flatten the 'subdefinitions' column
df3_normalized = pd.json_normalize(df3_exploded['details'], sep='_')

# Concatenate the flattened details with the original DataFrame
df3_result = pd.concat([df3_exploded[['word', 'translation', 'subword', 'type']], df3_normalized.explode('subdefinitions')], axis=1)
df3_result.head()
```
